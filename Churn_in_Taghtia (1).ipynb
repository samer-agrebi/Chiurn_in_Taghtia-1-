**I.INTRODUCTION**

* Qu'est-ce que le taux de désabonnement ?

  -Le taux de désabonnement est défini comme le moment où un client ou un abonné cesse de faire affaire avec une entreprise ou un service. Les clients de l'industrie des télécommunications peuvent choisir parmi une variété de fournisseurs de services et basculer activement entre eux. Dans ce marché hautement concurrentiel, l'industrie des télécommunications réalise un chiffre d'affaires annuel de 15 à 25 %.

  -La fidélisation personnalisée des clients est difficile car la plupart des entreprises ont un grand nombre de clients et ne peuvent pas consacrer beaucoup de temps à chaque client. Le coût est trop élevé pour compenser les revenus supplémentaires. Mais si une entreprise peut prédire à l'avance quels clients sont susceptibles de partir, elle peut concentrer ses efforts de fidélisation de la clientèle uniquement sur les clients « à haut risque ». L'objectif ultime est d'étendre sa portée et d'accroître la fidélité des clients. Le cœur du succès sur ce marché réside dans les clients eux-mêmes.

  -Le taux de désabonnement des clients est une mesure essentielle car il est beaucoup moins coûteux de fidéliser les clients existants que d'en acquérir de nouveaux.

  -Pour réduire l'attrition, les opérateurs doivent prévoir quels clients sont à haut risque d'attrition. La détection des premiers signes d'attrition potentielle commence par une compréhension globale des clients et de leurs interactions sur plusieurs canaux, y compris les visites en magasin/succursale, l'historique des achats de produits, les appels au service client, les transactions en ligne et les interactions sur les réseaux sociaux. 
  
   -Par conséquent, en s'attaquant au taux de désabonnement des clients, ces entreprises peuvent non seulement maintenir leur position sur le marché, mais aussi se développer. Plus ils ont de clients dans le réseau, plus les coûts de démarrage sont bas et plus le profit est élevé. 
   
   -En conclusion, l'objectif principal de la réussite commerciale est de réduire le taux de désabonnement des clients et de mettre en œuvre une stratégie de fidélisation efficace.

**VIII.DATA PREPROCESSING**

```python
data = data.drop(['Customer_Name'], axis = 1)
data.head()
```

```python
def object_to_int(dataframe_series):
    if dataframe_series.dtype=='object':
        dataframe_series = LabelEncoder().fit_transform(dataframe_series)
    return dataframe_series
```

```python
data = data.apply(lambda x: object_to_int(x))
data.head(10)
```

```python
plt.figure(figsize=(10,5))
print(data.corr()['Churn'].sort_values(ascending = False))
```

```python
X = data.drop(columns = ['Churn'])
y = data['Churn'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.30, random_state = 40, stratify=y)
```

```python
def distplot(feature, frame, color='springgreen'):
    plt.figure(figsize=(8,3))
    plt.title("Distribution for {}".format(feature))
    ax = sns.distplot(frame[feature], color= color)
num_cols = ["Tenure", 'Monthly_Charges', 'Total_Charges']
for feat in num_cols: distplot(feat, data)
```

* Étant donné que les caractéristiques numériques sont réparties sur différentes plages de valeurs, j'utiliserai un scalaire standard pour les réduire à la même plage.

*Standardizing numeric attributes*

```python
data_std = pd.DataFrame(StandardScaler().fit_transform(data[num_cols].astype('float64')),
                       columns=num_cols)
num_cols = ["Tenure", 'Monthly_Charges', 'Total_Charges']
for feat in num_cols: distplot(feat, data_std, color='mediumblue')
```

```python
# Divisez les colonnes en 3 catégories, une pour la normalisation, une pour l'encodage des étiquettes et une pour l'encodage à chaud

cat_cols_ohe =['Payment_Method', 'Contract', 'Internet_Service'] # those that need one-hot encoding
cat_cols_le = list(set(X_train.columns)- set(num_cols) - set(cat_cols_ohe)) #those that need label encoding
```

```python
scaler= StandardScaler()

X_train[num_cols] = scaler.fit_transform(X_train[num_cols])
X_test[num_cols] = scaler.transform(X_test[num_cols])
```

# IV.1.SVM

# IV.6.VotingClassifier

* Prédisons maintenant le modèle final basé sur la plus grande majorité des votes et vérifions son score.

```python
from sklearn.ensemble import VotingClassifier
clf1 = GradientBoostingClassifier()
clf2 = LogisticRegression()
clf3 = AdaBoostClassifier()
eclf1 = VotingClassifier(estimators=[('gbc', clf1), ('lr', clf2), ('abc', clf3)], voting='soft')
eclf1.fit(X_train, y_train)
predictions = eclf1.predict(X_test)
print("Final Accuracy Score ")
print(accuracy_score(y_test, predictions))
```

```python
print(classification_report(y_test, predictions))
```

```python
plt.figure(figsize=(4,3))
sns.heatmap(confusion_matrix(y_test, predictions),
                annot=True,fmt = "d",cmap='flag',linewidths=3)
    
plt.title("Final Confusion Matrix",fontsize=14)
plt.show()
```

* D'après la matrice de confusion, nous pouvons voir qu'il y a un total de 160 + 8 = 168 valeurs réelles de non-attrition, dont l'algorithme prédit 160 comme non-attrition et 8 comme des pertes. Alors que la valeur de désabonnement réelle est 29 + 26 = 55, l'algorithme prédit que 29 d'entre eux sont des valeurs de non-désabonnement et 26 d'entre eux sont des valeurs de désabonnement. Le taux de désabonnement des clients est définitivement mauvais pour le résultat net d'une entreprise. 

* Diverses stratégies peuvent être mises en œuvre pour éliminer le taux de désabonnement des clients. La meilleure façon d'éviter le désabonnement des clients est que les entreprises comprennent vraiment leurs clients. Cela comprend l'identification des clients à risque de désabonnement et l'amélioration de leur satisfaction. L'amélioration du service à la clientèle est bien sûr une priorité absolue pour résoudre ce problème.

* La fidélisation de la clientèle grâce à des expériences pertinentes et à un service spécialisé est une autre stratégie pour réduire le taux de désabonnement. Certaines entreprises sondent les clients qui se sont déjà désabonnés pour comprendre les raisons de leur départ afin d'adopter une approche proactive pour éviter le désabonnement futur de la clientèle.
